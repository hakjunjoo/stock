### google bigquery 기존 데이터 불러오기
- 기존 데이터 불러온 후 새로 받은 데이터와 결합
- 중복 제거

```python
import pandas_gbq

query1 = 'select * from yfinance.samsung order by Date'
query2 = 'select * from yfinance.sk order by Date'
query3 = 'select * from yfinance.naver order by Date'

samsung_gbq_df = pandas_gbq.read_gbq(query1, project_id='finance-420106')
sk_gbq_df = pandas_gbq.read_gbq(query2, project_id='finance-420106')
naver_gbq_df = pandas_gbq.read_gbq(query3, project_id='finance-420106')

display(samsung_gbq_df.head(3))
display(sk_gbq_df.head(3))
display(naver_gbq_df.head(3))
```

```python
# Date type 맞추기
# df : datetime, gbq_df : dbtime

samsung_gbq_df['Date'] = pd.to_datetime(samsung_gbq_df['Date'])
sk_gbq_df['Date'] = pd.to_datetime(sk_gbq_df['Date'])
naver_gbq_df['Date'] = pd.to_datetime(naver_gbq_df['Date'])
```

```python
# 데이터 결합 및 중복 제거
samsung_merge_df = pd.concat([samsung_df, samsung_gbq_df], axis=0).drop_duplicates(subset='Date')
sk_merge_df = pd.concat([sk_df, sk_gbq_df], axis=0).drop_duplicates(subset='Date')
naver_merge_df = pd.concat([naver_df, naver_gbq_df], axis=0).drop_duplicates(subset='Date')

# 데이터 정렬 및 인덱스 초기화
samsung_merge_df = samsung_merge_df.sort_values(by='Date').reset_index().drop(columns={'index'})
sk_merge_df = sk_merge_df.sort_values(by='Date').reset_index().drop(columns={'index'})
naver_merge_df = naver_merge_df.sort_values(by='Date').reset_index().drop(columns={'index'})
```

```python
# gbq로 저장
# 주의할 점 : column명이 숫자, 한글 시작이면 안된다. 띄어쓰기가 되어있는 경우도 제대로 업데이트가 안된다.
# if_exists 종류 : replace(덮어쓰기), append(합치기), fail(에러)

samsung_merge_df.to_gbq(destination_table='yfinance.samsung', project_id='finance-420106', if_exists='replace')
sk_merge_df.to_gbq(destination_table='yfinance.sk', project_id='finance-420106', if_exists='replace')
naver_merge_df.to_gbq(destination_table='yfinance.naver', project_id='finance-420106', if_exists='replace')
```
